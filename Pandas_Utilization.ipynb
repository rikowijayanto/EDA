```
#grouping the rank

df_rank = df.groupby(['rank'])
df_rank
```

```
#grouping rank based on count
df_rank.mean()
```

```
#group by salary
df_salary = df.groupby(['rank'])[['salary']].mean()
df_salary
```

```
#data filtering

df_filter = df[df['salary'] > 10000]
df_filter
```

## Slicing

```
#select particular columns
df_particular = df[['discipline', 'rank', 'salary', 'sex']]
df_particular
```

```
#select data in row 10 to 20 (data starting from 0)
df[9:19]
```

```
#select data in particular row and particular columns
df_part2 = df.loc[9:19, ['rank', 'salary']]
df_part2
```

```
#we can also use iloc if we use the index columns
df_part3 = df.iloc[9:19, [1, 6]]
df_part3
```

```
#iloc slicing
#first row
df.iloc[0]
```

```
#last row
df.iloc[-1]
```

```
#several rows
df.iloc[1:4]
```

```
#all rows and first two column
df.iloc[:, 0:2]
```

```
#row 1:7 and column 3-4
df.iloc[0:7, 2:5]
```

## Sorting

```
#sorting based

df_sort = df.sort_values(by='salary')
df_sort
```

```
#sorting based on salary and descending
df_sort1 = df.sort_values(by='salary', ascending=False)
df_sort1.head()
```

```
#sorting by two params and two ascending method
df_sort2 = df.sort_values(by=['rank', 'salary'], ascending=[True, False])
df_sort2
```
